**Respuesta:**
El parámetro alpha en la síntesis de voz con Tacotron controla la velocidad de habla y su valor produce los siguientes cambios:

* **alpha < 1.0**: Valores más bajos que 1.0 ralentizan la velocidad de habla.
Cuando alpha se acerca a cero, la síntesis será más lenta y se darán pausas más largas entre las palabras.
* **alpha = 1.0**: Un valor de 1.0 generalmente representa la velocidad de habla normal o predeterminada del modelo.
* **alpha > 1.0**: Valores más altos que 1.0 aceleran la velocidad de habla.
A medida que alpha aumenta, la síntesis se vuelve más rápida, con menos pausas entre las palabras.

Es importante tener en cuenta que los límites específicos pueden variar según el modelo Tacotron y las configuraciones específicas del sistema.
En este código, hemos probado con alpha=0.3 y alpha=1.2. Podríamos experimentar con valores más bajos (por ejemplo, entre 0.1 y 0.9) y valores más altos (por ejemplo, mayores a 1.2) para observar cómo afectan a la velocidad de habla. Esto nos permitirá entender mejor los límites y encontrar un valor que se ajuste a nuestras necesidades.

**Cuestion 2: Ver los límites del cambio de tono que el sistema puede generar**

**Respuesta:**

Estamos utilizando un pitch_function que multiplica la frecuencia fundamental por un factor de 7, lo que puede aumentar significativamente el tono de la voz sintetizada.

Los límites prácticos del cambio de tono dependerán del modelo específico que utilicemos y la calidad de la síntesis resultante. Debemos tener en cuenta los siguientes puntos:

* **Valores bajos de multiplicación**: Un pitch_function que multiplica la frecuencia por valores cercanos a 1.0 resultará en cambios de tono mínimos. La voz seguirá siendo reconocible y natural.
* **Multiplicación Moderada**: Valores moderados, como multiplicar por 1.5 o 2.0, generarán un cambio de tono perceptible pero aún dentro de límites razonables. La voz puede sonar más aguda pero no necesariamente distorsionada.
* **Multiplicación alta**: valores altos, como multiplicar por 5.0 o 7.0 como en el ejemplo 1, generarán cambios de tono más drásticos. Sin embargo, hay un límite en cuanto a la calidad y naturalidad de la voz, y valores demasiado altos pueden producir una voz sintética menos realista, que es lo que comprobamos en la primera pista de audio.

**Posibles distorsiones** (más allá de ciertos límites): A medida que aumentamos la multiplicación, se puede llegar a un punto en el que la síntesis se vuelve menos inteligible y más distorsionada. Valores extremadamente altos pueden hacer que la voz suene mecánica o robótica (ejemplo 1).
