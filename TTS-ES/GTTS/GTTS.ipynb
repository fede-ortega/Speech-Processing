```python
# Instalamos el programa
!pip install gtts
from gtts import gTTS
from IPython.display import Audio
```

PARÁMETROS QUE SE UTILIZAN

**language** : es el idioma en el que se quiere convertir el texto en habla

**dialecto_diferenciado** : es el dialecto específico que se utilizará: España= es, USA = us, Mexico = com.mx

**text** : es el texto a sintetizar

**slow_par** : puede ser False (por defecto) o True (pronunciación más lenta)





```python
# Introducir abajo los parámetros
language = 'es'
dialecto_diferenciado = 'es'
slow_par = False
```

```python

# Introduzca el texto abajo sustituyendo las palabras que están entre comillas

text= "En un lugar de la Mancha, de cuyo nombre no quiero acordarme, no ha mucho tiempo que vivía un hidalgo de los de lanza en astillero, adarga antigua, rocín flaco y galgo corredor. Una olla de algo más vaca que carnero, salpicón las más noches, duelos y quebrantos los sábados, lentejas los viernes, algún palomino de añadidura los domingos, consumían las tres partes de su hacienda"
```

```python
# Generación de la voz
speech = gTTS(text= text, lang= language, slow = slow_par, tld= dialecto_diferenciado)
# guardo la voz en un fichero
speech.save("Quijote_normal.mp3")
# reproduzco el fichero
Audio("Quijote_normal.mp3")

```

La pronunciación de números puede variar en diferentes contextos y regiones, y a veces los sistemas de TTS pueden no ajustarse completamente a las preferencias locales o contextuales. Se han ilustrado algunos ejemplos donde la pronunciación generada por un software de TTS podría no coincidir con las preferencias comunes:

1. **Números telefónicos y DNI**: La pronunciación de números telefónicos puede diferir. Algunos sistemas pueden pronunciar cada dígito individualmente o como un solo número, mientras que las personas tienden a agruparlos de manera más natural.

2. **Horas del día**: Dependiendo de la región, las horas pueden pronunciarse de manera diferente. Por ejemplo, "15:30" podría pronunciarse como "quince treinta" en algunos lugares o contextos formales, pero como "tres y media/treinta de la tarde" en otros.

3. **Fechas**: la pronunciación de las fechas puede variar. Por ejemplo, en el formato "23/05/2024", algunos sistemas de TTS podrían pronunciar "veintitrés del cinco de 2024" como "veintrés de mayo de 2024".

4. **Números decimales**: La pronunciación de números decimales puede variar, especialmente en términos de la separación entre la parte entera y la decimal. Por ejemplo, el número "3,14" podría pronunciarse como "tres coma catorce" o "tres con catorce".

G.   De manera libre, buscar diferencias entre los dialectos de español diferenciados, España=es, USA= us y  Mejico= com.mx

**Respuesta:**
Las variaciones más significativas a nivel fonético entre los tres dialectos son las siguientes:

1. **Seseo y ceceo**: En España puede haber distinción entre "s" y "z" y "c" (ceceo), mientras que en el dialecto de México y Estados Unidos, ambos se pronuncian de la misma manera (seseo).

2. **Aspiración de /s/**: En algunas regiones de España, especialmente al final de las sílabas, la /s/ puede ser aspirada o incluso omitida. En América Latina y México, la /s/ suele mantenerse al final de las sílabas. En este caso, el español es de la variedad castellana y tampoco observamos aspiración.
3. **Yeísmo**: En España, algunas regiones de Castilla y León mantienen la distinción entre "ll" y "y", mientras que en otras, ambos sonidos se pronuncian de la misma manera (yeísmo). En América Latina y México, el yeísmo es más común, donde "ll" y "y" se pronuncian de manera similar.
4. **Vocalización de /r/**: En América Latina y México, la /r/ al final de las sílabas tiende a mantener su sonido más fuerte.
5. **Prosodia**: La entonación y el ritmo del habla pueden variar, y esto a menudo es más una cuestión de acento regional que de fonética precisa.
